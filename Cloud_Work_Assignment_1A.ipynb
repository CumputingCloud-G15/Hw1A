```python
#@title tirgul3 - ex

# mount google drive
from google.colab import drive
drive.mount('/content/drive')

from ssl import Options
import json
import ipywidgets as widgets
from IPython.display import display, HTML
from typing import Text


with open("/content/drive/MyDrive/Colab Notebooks/students_updated.json") as us:
  file_cont = us.read()
parsed_json = json.loads(file_cont)

class Student:
  def __init__(self, first_name, last_name, email, courses, link):
    self.first_name = first_name
    self.last_name = last_name
    self.email = email
    self.courses = courses
    self.link =link

students = []


for obj in parsed_json:
  students.append(Student(obj["first_name"],
                     obj["last_name"],
                     obj["email"], 
                     obj["courses"],
                     obj["link"]))


# building array of students from json file->string->object in array

student_Name = widgets.Dropdown(
    options = [s.first_name for s in students],
    layout = {'width': 'max-content'},
    value = None,
    placeholder = "John",
    description = 'First Name:',
)

last_Name = widgets.Text(
    value = '',
    placeholder='Snow',
    description='Last name:',
    disabled=True,
)

email = widgets.Text(
    value = '',
    placeholder='JSnowBack2Life@Winterfall.jrr.com',
    description='Email:',
    layout = {'width': '350px'},
    disabled=True,
)

coursesTxt = widgets.Text(
    value = '',
    placeholder='Guarding the wall 101',
    layout = {'width': '850px'},
    description = "Courses: ",
    disabled=True,
)

txtName = widgets.HTML(
    value = "<h1>welcome to our form, choose one of the students below:</h1>",
    disabled=True,
)

button = widgets.Button(
    description='Save data',
    disabled=True,
    button_style='',
    tooltip='Save favourite course of student',
)
link_url = widgets.HTML(
    value=None
)

tvShowText = widgets.Text(
    value = '',
    layout = {'width': '450px'},
    description = "Fav. show: ",
    disabled=True,
)

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
      for student in students:
        if(change['new'] == student.first_name):
          last_Name.value = student.last_name
          email.value = student.email
          link_url.value = f"<a href='{student.link}' target='_blank'><h3>Interesting link(Click me!)</h3></a>"
          coursesTxt.value = " | ".join(student.courses)
          tvShowText.disabled = False
          tvShowText.value = ''
          button.disabled = True
          break

def on_show_change(change):
  if change['type'] == 'change' and change['name'] == 'value':
    if(change['new'] != ''):
      button.disabled = False
    else:
      button.disabled = True

def on_save_click(b):
    for obj in parsed_json:
      if obj["first_name"] == student_Name.value:
        newData = {"fav_show": tvShowText.value}
        obj.update(newData)

    json_object = json.dumps(parsed_json, indent=4)
 
    # Writing to sample.json
    with open("/content/drive/MyDrive/Colab Notebooks/students_updated.json", "w") as outfile:
        outfile.write(json_object)

# insorting all widgets to a vbox and chaning its color
vbox = widgets.VBox([txtName, student_Name, last_Name, email, coursesTxt, tvShowText, link_url, button])
display(HTML("""
<style>
    .my-vbox {
        background-color: darkblue;
    }
</style>
"""))
vbox.add_class("my-vbox")
display(vbox)

# attaching the widgets to their functions
student_Name.observe(on_change)
tvShowText.observe(on_show_change)
button.on_click(on_save_click)



```
